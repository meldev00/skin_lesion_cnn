> This notebook is dedicated to playing around with manipulating and augmenting image data. 

```python
import pandas as pd
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
import PIL
from PIL import Image
from matplotlib import image
import matplotlib.pyplot as plt
```

```python
fig1 = plt.figure()

vis_im  = Image.open("../skin-cancer-mnist-ham10000/image_data_test/type_0/ISIC_0024843.jpg")
vis_im

fig1.savefig("../images/type0_sample.png")
```

```python
im = image.imread("../skin-cancer-mnist-ham10000/image_data_test/type_0/ISIC_0024843.jpg")
```

```python
imgen = ImageDataGenerator(
        rescale = 1./255,
        rotation_range = 360,
        width_shift_range = 0.2,
        height_shift_range = 0.2,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True,
        vertical_flip = True
)
```

```python
new_im = imgen.apply_transform(im, {"theta": 50,
                                    "flip_vertical": True})
```

```python
Image.fromarray(new_im)
```

```python
im_dict = imgen.get_random_transform((48,48,3))
```

```python
new_im2 = imgen.apply_transform(im, im_dict)
```

```python
Image.fromarray(new_im2)
```

> The "apply transform" and "get random transform will be useful for creating more augmented data.
